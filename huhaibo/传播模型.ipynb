## Diffusion in Social Networks
### Contagion
People adopt when they come in contact with others who have already adopted; that is, innovations spread much like epidemics.
### Social influence 
People adopt when enough other people in the group have adopted; that is, innovations spread by a conformity motive.
### Social learning 
People adopt once they see enough empirical evidence to convince them that the innovation is worth adopting, where the evidence is generated by the outcomes among prior adopters. Individuals may adopt at different times due to differences in their prior beliefs, amount of information gathered, and idiosyncratic costs.

H P Young. Innovation diffusion in heterogeneous populations: Contagion, social influence, and social learning. Am Econ Rev, 2009, 99: 1899-1924.

## 经典的信息扩散模型

### 阈值模型

* 社会学家Granovetter（1943-）提出了阈值模型(threshold model)，模型指出，一个人在群体行为中是否采取行动，不仅仅取决于自身，更取决于周边有多少跟自己想法类似的人(like-minded people)。只有跟自己想法类似的人的数量或比例超过某个阈值，该人才会采取行动。
* 经济学家Schelling（1921–2016）提出了隔离动态模型(dynamic models of segregation）。

之后，越来越多的阈值模型的变体得到了广泛的研究。

阈值模型的一个简单的例子是线性阈值模型：
* 社会网络表示为一个有向图$G=(V, E)$，网络中的个体有两种初始状态：激活状态和非激活状态。
* 在线性阈值模型中，某个节点一旦被激活，则将一直保持为激活状态，直到信息传播结束。
* 我们为每个节点$v$赋一个在[0, 1]内均匀分布随机抽取的激活阈值$\theta_v$，$\theta_v$反映的是节点$v$被邻居节点激活的难易程度，$\theta_v$越大表示节点$v$越难被其邻居节点激活，反之，$\theta_v$越小表示节点$v$越容易被其邻居节点激活。

* 假设结点$u$和$v$是$V$中的两个节点，即$u,v \in V$，节点$u$为激活状态，而节点$v$为非激活状态，有向边$(u, v)$上的权重表示为$w_{uv}$，则$w_{uv}$表示节点$u$对出边邻居节点(out neighbouring nodes)$v$的影响程度。
* 假设最初种子节点集合为$S_0$，在信息扩散过程中，第$t$步激活的节点集合为$S_t$(包含从第0步到第$t$步激活的所有节点)，那么在第$t+1$步，对于任何一个节点$v \in V\backslash S_t$，若所有处于激活状态的邻居节点对节点$v$的影响力总和大于等于节点$v$的阈值，即:
$$ \sum\limits_{u \in S_t \cap N_{{\rm{in}}} (v)} {w_{uv}  \ge \theta _v }  $$
则节点$v$被激活，否则节点$v$不被激活。
其中，$V\backslash S_t$表示节点集合$V$除去节点集合$S_t$剩余的节点集合，$N_{{\rm{in}}} (v)$表示节点$v$的入边邻居节点集合，$S_t \cap N_{{\rm{in}}} (v)$则表示第$t$步时节点$v$处于激活状态的入边邻居节点集合。
* 第$t+1$步激活的节点加入到$S_t$，形成$S_{t+1}$，重复这个信息传播过程，直到没有新的节点被激活。

* 阈值模型的特点是存在一个阈值，不管该阈值是邻居节点的比例$q$还是与个人相关的一个数值$\theta_v$，只有当与邻居节点的状态相关的某些变量超过该阈值时，该节点才会被激活。
* 该模型适用于这样的场合：某项新技术，某个体采纳该新技术的邻居对个体的影响必须达到某一程度，该个体才会采纳这种新技术。

### 级联模型

* Goldenberg等在营销背景下提出了扩散的级联模型。在该模型中，当一个个体变得活跃后会以一定的概率激活它邻居中的每一个不活跃个体。
* 一个简单的例子是独立级联模型，在该模型中，每个初始激活节点会产生自己独立的扩散级联，级联之间互相独立、互不干扰。模型中一个个体被一个最近活跃的邻居激活的概率独立于过去试图激活他/她的邻居的集合。

* 社会网络表示为一个有向图$G=(V，E)$，网络中的节点有两种初始状态：激活状态和非激活状态。
* 与线性阈值模型不同的是，在独立级联模型中，一个节点$u$一旦在第$t$步被激活，它将在第$t+1$步去激活它的邻居结点$v$，之后便失去激活能力，即每个被激活的节点只有一次机会去激活它的邻居节点。
* 假设种子节点为$S_0$，节点$v$被节点$u$激活的概率为$p_{uv}$。则独立级联模型按照如下的扩散规则进行传播：假设第$t-1$步处于激活状态的结点集合为$S_{t-1}$，第$t$步处于激活状态的节点集合为$S_t$，则在第$t+1$步时，每个节点$u \in N_{{\rm{in}}} (v) \cap (S_t \backslash S_{t - 1} )$以概率$p_{uv}$去尝试激活它的出边邻居节点$v$，其中$N_{{\rm{in}}} (v)$表示节点$v$的入边邻居节点集合。若成功，则节点$v$被激活，否则，在后面的传播过程中，节点$u$不再激活节点$v$。重复执行这一步骤，直到社会网络中再没有节点被激活。


* 也就是说，在时间$t$，每个当前激活的节点$u$都会以一定概率$p_{uv}$去激活它的每个不活跃邻居节点$v$。
* 如果在时刻$t$，节点$v$的多个邻居节点要激活它，这些邻居节点会随机排列地去尝试激活，所有激活尝试都在时刻$t$内完成。
* 无论邻居节点$u$是否成功，在随后的时刻都不能再去激活$v$。如果$v$在$t$时刻被激活，那么该节点会在$t+1$时刻去激活它的邻居们。

### 传染病模型

* 传染病传播和信息扩散具有很强的相似性，因此传染病模型可以应用到信息扩散或情绪扩散研究中。
* 基本的传染病模型包括SIS和SIR两种。在SIS模型中，节点有两种状态：易感状态和已感状态。在时刻$t$，如果节点$v$周围有一个或者多个已受感染的节点，那么节点$v$会以概率$p_i$变成已感状态，这里的$p_i$通常是固定的，不随用户关系的变化而变化，一段时间之后，处于已感状态的节点$v$又以概率$p_r$重新回归易感状态。
* 在SIR模型中，节点除了有上述两种状态外，还有第3种状态——免疫状态，即用户受到感染被治愈后具有了免疫性，不再被感染，也不会感染其它节点。

* 应用到信息扩散场景中：易感状态表示用户尚未知晓信息，已感状态表示用户已收到信息并将信息发给好友，免疫状态则表示用户虽已收到信息但认为该信息不具价值或对该信息不感兴趣因此不会发给好友，这些用户成为信息扩散树中的叶子节点。
* 通过不同状态的节点间的相互作用，节点的状态会发生变化，信息从而在整个网络中扩散。

下面给出网络上SIS模型的代码（SIS-model.py，取自Introduction to the Modeling and Analysis of
Complex Systems by Hiroki Sayama，代码可从http://bingweb.binghamton.edu/~sayama/textbook/下载）:

```python
import matplotlib
matplotlib.use('TkAgg')
from pylab import *
import networkx as nx
import random as rd
```

首先，初始化每个节点的状态，用0表示易感状态，1表示已感状态。

```python
def initialize():
    global g
    g = nx.karate_club_graph() # 以空手道俱乐部网络为例
    g.pos = nx.spring_layout(g)
    for i in g.nodes_iter():
        g.node[i]['state'] = 1 if random() < .5 else 0
```

可视化网络：

```python
def observe():
    global g
    cla()
    nx.draw(g, cmap = cm.binary, vmin = 0, vmax = 1, 
            node_color = [g.node[i]['state'] for i in g.nodes_iter()],
            pos = g.pos)     
```

节点状态更新如下所示：

```python
p_i = 0.5 # 感染概率
p_r = 0.5 # 恢复概率
def update():
    global g
    a = rd.choice(g.nodes())
    if g.node[a]['state'] == 0: # 如果a节点为易感节点
        b = rd.choice(g.neighbors(a))
        if g.node[b]['state'] == 1: # 若邻居节点b被感染
            g.node[a]['state'] = 1 if random() < p_i else 0
    else: # 如果a节点为已感节点
        g.node[a]['state'] = 0 if random() < p_r else 1         
```

## 数据驱动的信息扩散模型

### 分支过程模型

* 在新媒体时代，由于可以获得大规模的信息扩散数据，在数据驱动的信息扩散模型的构建与分析方面也取得了一定的进展。
* 很多此类模型基于分支过程，在这里我们考虑离散时间分支过程，最简单的模型，即Galton-Watson分支过程模型如下图所示。
* 在该过程中，一个信息源将信息发给好友，这些好友又将信息发给他们的好友，依此进行下去，发给的好友的数量往往满足特定的分布。
<img src="http://localhost:8888/tree/figureModel/fenzhi.png"style="width:500px;float:middle">

* 分支过程的一个基本属性是，如果其某一代的个体没有人再扩散消息，则消息就会消失。因此该模型实际上只有两种信息扩散的可能性：没有人再继续传播信息，信息因此在经过有限的传播步骤后消失；或者人们继续传播信息，这个过程一直无限的在社会网络上持续下去。
* 假定收到信息后个体转发该信息的概率为$p$，个体将信息转发给的好友数量的均值为$k$，则$R_0 = pk$即是基本再生数。若$R_0 \le 1$信息将在有限的时间步后以概率1消失，若$R_0 > 1$则信息持续在每一步以大于0的概率至少传给一个人。

D Easley, J Kleinberg. 网络、群体与市场. 清华大学出版社.

* 从基本再生数可知信息扩散至少跟两个因素有关，一是网络结构，二是人们传播信息的意愿。
* 我们假设第$n$个时间步传播到的个体的数量为$Z_n$，初始时信息源节点或种子节点的数量为$Z_0$，并假设个体$i$将信息传播给了$\xi _i $个个体，则第$n+1$个时间步传播到的个体的数量为$Z_{n + 1}  = \sum\nolimits_{i = 1}^{Z_n } {\xi _i } $。假设$\xi _i $是独立同分布随机变量，其均值为$k$，则$Z_n$的均值为E[$Z_n$]=$k^n$E[$Z_0$]。因此我们发现当$k<1$，$n \to \infty $时平均规模以指数衰减并趋近于0，而当$k>1$时则以指数速度增长并趋近于无穷。

假设每个个体将信息传播给了$\xi$位其他用户，$\xi$是随机变量，设初始时种子节点数量为1即$Z_0=1$，E($\xi$)=$\mu$，如果$\mu<1$,信息将在有限时间步后消失，知晓信息的个体的数量为一有限值，其均值为E($Y$)=$(1-\mu)^{-1}$，更一般的E($Y$)=$Z_0(1-\mu)^{-1}$。

### 基于信息亲和性的模型

* 在微博中，影响用户转发微博的因素是多方面的，如微博发布者特征、微博文本特征、发布者与其粉丝的社交关系特征等。
* 以用户的转发行为为因变量，各种影响因素为自变量，逻辑回归分析得到的每个因素的系数分布如下图所示。
* 我们发现，用户与其粉丝的交互强度、用户与其粉丝的兴趣相似度、微博原创性以及微博主题和用户主题相似度对微博的转发有显著的正向影响。
* 信息亲和性指的便是用户对消息的感兴趣程度，即用户倾向于转发他们感兴趣的信息而忽略掉他们不感兴趣的信息。
<img src="http://localhost:8888/tree/figureModel/qinhe.png"style="width:800px;float:middle">

基于信息亲和性的模型中，每个节点可被赋予一个亲和性值来刻画该节点对消息的亲和性。在每一个时间步，网络节点可以是下面三种状态中的任一种：S状态即节点没有收到消息，I状态即节点正在传播消息，R状态即节点不再传播消息。

* 不像SIR模型，信息亲和性模型不用全局概率来调控节点状态转移。与之相反，节点转发信息的决策取决于用户对信息的兴趣程度（即用户传播信息的倾向程度）和信息自身的吸引力（即适合传播的概率或适配性）之间的相互作用。
* 每个节点的亲和性$a_i \in [0,1]$假定服从特定的概率分布，表示节点参与信息扩散的倾向，触发节点传播行为的信息适配性用信息的亲和性阈值$A_T \in [0,1]$表示。
* S状态的节点收到消息后，只有当他们的亲和性大于消息阈值时（$a_i>A_T$）才能成为I状态，否则成为R状态。所以本质上该模型融合了SIR模型和阈值模型。

### Voter model投票者模型

* 投票者模型是一种舆论或观点动力学模型，可用于研究社会网络中舆论的扩散及相互作用。
* 在舆论扩散模型中，个体的观点可用数字来表示，常见的是二元观点或选择，如生活中常见的PC–Macintosh、Windows–Linux、coffee–tea、买-卖等。
* 除了二元观点，现实中还存在可选择的观点数超过2的情况，如购买某种品牌的商品或多党制国家中的选举等。在这两种情况中，个体的观点$s_i$均可用有限数量的数字（一般为整数）来表示，如$s_i  = 1,2, \ldots ,I\;(I \ge 2)$,，因而称之为离散观点。

* 投票者模型最初用来研究生物物种之间的竞争，但它在解释某些社会、经济现象方面非常有效。
* 投票者模型的定义非常简单：每个个体的观点是二元变量$s =  \pm 1$，在每一时间步，随机选择一个个体$i$和它的一个邻居节点$j$，并更新其观点为$s_i=s_j$，即个体模仿它的邻居的观点。
* 在该模型中，一般情况下，所有个体的观点相同或一致是稳定状态（吸收态）。

网络上投票者模型的代码如下（voter-model.py，取自Introduction to the Modeling and Analysis of Complex Systems by Hiroki Sayama，可从http://bingweb.binghamton.edu/~sayama/textbook/下载）。

```python
import matplotlib
matplotlib.use('TkAgg')
from pylab import *
import networkx as nx
import random as rd
```

初始化每个节点的状态，共有0和1两种状态。

```python
def initialize():
    global g
    g = nx.karate_club_graph() # 以空手道俱乐部网络为例
    g.pos = nx.spring_layout(g)
    for i in g.nodes_iter():
        g.node[i]['state'] = 1 if random() < .5 else 0
```

可视化网络：

```python
def observe():
    global g
    cla()
    nx.draw(g, cmap = cm.binary, vmin = 0, vmax = 1, 
            node_color = [g.node[i]['state'] for i in g.nodes_iter()],
            pos = g.pos)          
```

节点状态更新如下所示：

```python
def update():
    global g
    listener = rd.choice(g.nodes()) # 随机选择一个节点
    speaker = rd.choice(g.neighbors(listener)) # 随机选择该节点的一个邻居
    g.node[listener]['state'] = g.node[speaker]['state'] # 状态更新
```

### Axelrod model
* In Axelrod model, an agent is represented by a string of cultural features, where each feature can adopt a certain number of distinct traits. 
* The interaction between any two agents takes place with probability proportional to their cultural similarity, i.e., proportional to the percentage of their features that have the identical trait. 

* When interacting, one agent influences another agent causing the latter to adopt the former’s trait on a feature randomly chosen from those that they do not share. 
* The basic idea is that individuals who are similar to each other are likely to interact and then become even more similar. This process of social influence is implemented by assuming that the probability of interaction between two neighboring individuals is equal to their degree of similarity.


The local evolution rule consists of the following steps.
1. Select at random an agent to be active and one of its neighbors.
2. Select at random a feature on which the active agent and its neighbor differ, and change the active agent’s trait on this feature to the neighbor’s trait on this feature with a probability equal to the degree of similarity between the two agents. If the two agents have a degree of similarity equal to 100%, nothing happens.

* The process continues until no cultural change can occur. This happens when every pair of neighboring agents has cultures that are either identical or completely different. 
* Two final possible states are possible: only one cultural region is obtained or multiple cultures are obtained separated by a boundary.

### DeGroot Model
* 观点形成和共识达成的网络交互作用的开创性模型归功于DeGroot。
* 社会中的每个人对某一特定事物都有一个最初的观点，它们可用$n$维概率向量${\bf{p}}(0)=(p_1(0), …, p_n(0))$表示，其中$p_i(0)$位于区间[0, 1]内，它可视为某个特定的表述成立的可能性，或者是给定产品的质量，或是个人参加到特定活动中的概率。
* 个体间交互作用的模式通过权重和有向的$n*n$非负矩阵来表示。


特别的，设${\bf{W}}$是一个行随机矩阵，即每一行各项和为1。$w_{ij}$表示个体$i$在下一时间步的观点中，赋予个体$j$当前观点的权重或信任度，或$j$对$i$的影响权重。

个体观点随着时间的变化而更新，即
$${\bf{p}}(t)  =  {\bf{Wp}}(t - 1)  =  {\bf{W}}^t {\bf{p}}(0)$$

### 社会影响
$${\bf{s}}={\bf{sW}}$$
${\bf{s}}$为每个个体的极限影响，是${\bf{W}}$的一个左侧单位特征向量。DeGroot模型为一些以特征向量为基础的中心性和影响指标（如Pagerank）提供了一个明确基础。


### Friedkin模型
* Friedkin模型类似于DeGroot模型，因此有时也将两者合称为DeGroot-Friedkin模型。
* 它与DeGroot模型不同的地方在于：不仅考虑了邻居个体的影响，也考虑了自身初始观点的影响。
$${\bf{p}}(t)  = {\bf{A}} {\bf{Wp}}(t - 1) +({\bf{I}}-{\bf{A}}){\bf{p}}(0)$$
其中${\bf{A}}=[a_{11}, a_{22}, ..., a_{nn}]$，$0 \le a_{ii} \le 1$，${\bf{I}}=[1, 1, ..., 1]$。
因此，若$({\bf{I}}-{\bf{A}})={\bf{0}}$，则Friedkin模型退化为DeGroot模型。

### 有界信任模型
* 很多模型中，不管个体的状态如何，每个个体都可与他/她的邻居相互作用，而实际上，现实生活中，人与人之间的交流往往仅局限于相互之间观念相差不是很大的群体，相差太大会导致没有共同语言，人际交流的这一特性被称为有界信任（Bounded Confidence）。
* 在模型中，通常引入一实数$\varepsilon $，一个观点为$s$的个体仅能与观点在区间$[s - \varepsilon ,\;s + \varepsilon ]$内的邻居个体相互作用，$\varepsilon $被称为不确定度或容忍度。

### Deffuant模型
Deffuant模型为Bounded Confidence模型，其规则如下：

初始时网络上的每个个体$i$被赋予一个在[0, 1]区间内随机选择的观点$s_i$。在每一时间步，随机选择的一个个体与他/她的一个邻居相互讨论。令$i$和$j$为$t$时刻相互交流的两个个体，其观点分别为$s_i (t)$和$s_j (t)$。若$s_i (t)$和$s_j (t)$的差值的绝对值超过$\varepsilon $，个体观点不发生变化，而若$\left| {s_i (t) - s_j (t)} \right| < \varepsilon $，则
$$\left\{ \begin{array}{l}
 s_i (t + 1) = s_i (t) + \mu \left[ {s_j (t) - s_i (t)} \right] \\ 
 s_j (t + 1) = s_j (t) + \mu \left[ {s_i (t) - s_j (t)} \right] \\ 
 \end{array} \right.$$
$\mu$称为收敛参数，其值在[0, 1/2]区间。

Deffuant模型基于一种妥协策略，经过一轮交流或辩论后，相互作用的两个个体的观点以相对量$\mu$互相接近对方。若$\mu = 1/2$，两个个体的观点将收敛至二者辩论前的均值。对于任意的$\varepsilon $和$\mu$，相互作用前后两个个体的观点的均值均相同，因而整个社会网络中所有个体的观点的均值1/2是Deffuant模型的一个全局不变量。

### Hegselmann-Krause（HK）模型
HK模型非常类似于Deffuant模型，不同之处在于更新规则。
* 在Deffuant模型中，个体$i$仅与满足条件的一个邻居相互作用，而在HK模型中，$i$同时与满足条件的所有邻居相互作用。
* Deffuant模型适合描述大群体中的舆论动力学，个体只是小规模地，一般是两两之间，进行观念互动，而HK模型则适于描述比较正式的会议，在同一时间，很多人可进行有效的交流。

HK模型中个体$i$的观点在时间$t$的更新规则为：
$$s_i (t + 1) = \sum\limits_{j:{\rm{|}}s_i (t) - s_j (t){\rm{|}} < \varepsilon } {a_{ij} s_j (t)} {\rm{/}}\sum\limits_{j:{\rm{|}}s_i (t) - s_j (t){\rm{|}} < \varepsilon } {a_{ij} } $$
其中$a_{ij}$表示邻接矩阵。可见$i$取他/她的符合条件的邻居状态的均值。若取消有界信任的限制，HK模型将类似于DeGroot模型。

HK模型完全由$\varepsilon $决定，而在Deffuant模型中，除了$\varepsilon $，我们还必须确定收敛参数$\mu$。

HK模型演化的结果类似于Deffuant模型，随着$\varepsilon $的增大，最终剩余的观点簇的数量将减少。特别地，当$\varepsilon $超过某一阈值$\varepsilon_c$时，网络中将仅存在一个观点簇，即所有人的观点达到一致。
