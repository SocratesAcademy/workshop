## 1.2 样式及参数
NetworkX提供了一系列样式参数，可以用来修饰和美化图形，达到我们想要的效果。
常用的参数包括：

- `node_size`:  指定节点的尺寸大小(默认是300，单位未知，就是上图中那么大的点)
- `node_color`:  指定节点的颜色 (默认是红色，可以用字符串简单标识颜色，例如'r'为红色，'b'为绿色等，具体可查看手册)
- `node_shape`:  节点的形状（默认是圆形，用字符串'o'标识，具体可查看手册）
- `alpha`: 透明度 (默认是1.0，不透明，0为完全透明)
- `width`: 边的宽度 (默认为1.0)
- `edge_color`: 边的颜色(默认为黑色)
- `style`: 边的样式(默认为实现，可选： solid|dashed|dotted,dashdot)
- `with_labels`: 节点是否带标签（默认为True）
- `font_size`: 节点标签字体大小 (默认为12)
- `font_color`: 节点标签字体颜色（默认为黑色）

对1.1节中生成的网络图形G逐项进行参数调整：

```python
#node_size:  指定节点的尺寸大小(默认是300) 
nx.draw(G,node_size = 30) 
```

```python
#node_color:  指定节点的颜色 (默认是红色，可以用字符串简单标识颜色，例如'r'为红色，'b'为绿色等)
nx.draw(G, node_color = 'b') 
```

```python
#node_color: 节点颜色使用[0,1]范围内的三个元素的元组来表示，例如(1.0,0.0,0.0)表示红色，(0.0,1.0,0.0)表示绿色，(0.0,1.0,0.0)表示蓝色
nx.draw(G, node_color = (0.0,1.0,0.0)) 
```

```python
#node_color: 节点颜色可由用户随机设定
colors = [nr.rand(),nr.rand(),nr.rand()]   # 节点的颜色由随机生成的三个元素组合决定，如[0.5004917205978098, 0.5857564995520697, 0.9038947318144]
nx.draw(G, node_color=colors)
```

```python
#node_color: 每个节点的颜色可以不同，且节点颜色可由用户随机设定
colors = [(nr.rand(),nr.rand(),nr.rand()) for i in range(10)]   # 节点的颜色由随机生成的三个元素组合决定，其中range()中的数值决定颜色种类的数目
nx.draw(G, node_color=colors)
```

```python
#node_color: 每个节点的颜色可以不同，且节点颜色可为黑白渐近色

fig = figure(figsize=(15,5))
ax1 = fig.add_subplot(131)
nx.draw(G, node_color = str(0.0))   # 0.0代表黑色

ax2 = fig.add_subplot(132)
nx.draw(G, node_color = str(1.0))   # 1.0代表白色

ax3 = fig.add_subplot(133)
nx.draw(G, node_color = str(0.5))   # 0.5代表灰度值为5
```

```python
#node_shape:  节点的形状（默认是圆形，用字符串'o'标识）
nx.draw(G, node_shape = 's')
```

```python
#alpha: 透明度 (默认是1.0，不透明) 
nx.draw(G, alpha = 0.4)
```

```python
#edge_color: 边的颜色(默认为黑色)
nx.draw(G, edge_color ='r')
```

```python
#style: 边的样式(默认为实线，可选： solid|dashed|dotted,dashdot)
nx.draw(G, style = 'dashed')
```

```python
#with_labels: 节点是否带标签（默认为False）
nx.draw(G, with_labels=True)
```

```python
# 随机设置节点的颜色
n=G.number_of_nodes()
colors = [(nr.rand()) for i in range(n)]       # 每个节点的颜色随机设置，颜色数值用[0,1]范围内的单个小数表示
nx.draw(G,with_labels=True, node_size = 400,node_color=colors)
```

```python
# 用三元组表示节点的颜色，使色彩更丰富
n=G.number_of_nodes()
colors = [(nr.rand(),nr.rand(),nr.rand()) for i in range(n)]   # 每个节点的颜色随机设置，颜色数值用[0,1]范围内的3个小数表示
nx.draw(G,with_labels=True, node_size = 400,node_color=colors)
```

```python
#font_size: 节点标签字体大小 (默认为12)
nx.draw(G, with_labels=True, font_size=18)
```

```python
#font_color: 节点标签字体颜色（默认为黑色）
nx.draw(G, with_labels=True, font_size=16, font_color='g')
```

```python
# 根据节点随机属性信息值大小和颜色
n=G.number_of_nodes()
dates = [(nr.rand()) for i in range(n)]     # n个节点的属性信息值dates随机初始化
colors=dates                                 # 根据节点的属性信息值dates赋值节点颜色
sizes=[(dates[i]*800) for i in range(n)]    # 根据节点的属性信息值dates赋值节点大小
nx.draw(G,with_labels=True, node_size = sizes,node_color=colors)
```

```python
# 根据节点的随机属性信息值dates赋值节点颜色，以黑白灰度显示
nx.draw(G,with_labels=True,cmap=plt.cm.Greys,node_size=sizes,node_color=colors)
```

```python
# 根据网络节点的指定属性值，以灰度图调整节点的大小和颜色，显示相应的灰度分布标尺
n=G.number_of_nodes()
data= np.random.random((n,1))
print 'The data is',data

#获取nx.draw函数中G.nodes()节点默认排列顺序
node = G.nodes()
index = []
for i in node:
    index.append(int(i))

# 按照默认节点排列顺序index生成节点属性列表
dates_new = []    
for i in range(n):
    dates_new.append(data[index[i]]) 

# 根据节点的指定属性信息值dates赋值节点颜色和大小   
sizes=[x*800+100 for x in dates_new]
colors=[x for x in dates_new]

# 画网络节点图及标尺，以黑白灰度显示 
fig, ax = plt.subplots()
heatmap = ax.pcolor(data, cmap=plt.cm.Greys, vmin=0.0, vmax=1.0)
plt.subplot(1,1,1)
nx.draw(G,with_labels=True,cmap=plt.cm.Greys,node_size=sizes,node_color=colors)
cbar = plt.colorbar(heatmap)  # 添加标尺
```

```python
# 根据网络节点的指定属性值，以光谱彩色图调整节点的大小和颜色，显示相应的光谱彩色分布标尺
n=G.number_of_nodes()
data= np.random.random((n,1))
print 'The data is',data

#获取nx.draw函数中G.nodes()节点默认排列顺序
node = G.nodes()
index = []
for i in node:
    index.append(int(i))

# 按照默认节点排列顺序index生成节点属性列表
dates_new = []    
for i in range(n):
    dates_new.append(data[index[i]]) 

# 根据节点的指定属性信息值dates赋值节点颜色和大小   
sizes=[x*800+300 for x in dates_new]
colors=[x for x in dates_new]

# 画网络节点图及标尺，以黑白灰度显示 
fig, ax = plt.subplots()
heatmap = ax.pcolor(data, vmin=0.0, vmax=1.0)
plt.subplot(1,1,1)
nx.draw(G,with_labels=True,node_size=sizes,node_color=colors)
cbar = plt.colorbar(heatmap)  # 添加标尺
```

## 1.3 节点布局
节点布局指定网络中节点排列的形式

```python
pos = nx.circular_layout(G)      #circular_layout：节点在一个圆环上均匀分布         
nx.draw(G,pos)
```

```python
pos = nx.random_layout(G)       #random_layout：节点随机分布    
nx.draw(G,pos)
```

```python
pos = nx.spring_layout(G)       #spring_layout： 用Fruchterman-Reingold算法排列节点，默认布局方式      
nx.draw(G,pos)
```

```python
pos = nx.spectral_layout(G)      #spectral_layout：根据图的拉普拉斯特征向量排列节点    
nx.draw(G,pos)
```

## 1.5 读取网络数据

```python
# 连边对数据可以直接读取来生成网络
# 注意：读取网络数据文件"BA_network.txt"必须保证该文件已经上传至指定路径
G = nx.read_edgelist("BA_network.txt")
G = G.to_undirected() 
```
